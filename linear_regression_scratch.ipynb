# Learning Linear Regression from Scratch
## Introduction
Linear Regression is a fundamental supervised learning algorithm used for predicting continuous values. It models the relationship between an independent variable (X) and a dependent variable (Y) using a linear equation.

### Linear Equation:
$$ y = mx + c $$
For multiple features:
$$ Y = XW + b $$

## 1. Cost Function
The cost function measures the error of predictions. We use **Mean Squared Error (MSE):**
$$ J(W, b) = \frac{1}{2m} \sum_{i=1}^{m} (h(x_i) - y_i)^2 $$

## 2. Gradient Descent Algorithm
To minimize the cost function, we use **Gradient Descent**:
$$ W := W - \alpha \frac{\partial J}{\partial W} $$
$$ b := b - \alpha \frac{\partial J}{\partial b} $$

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate Sample Data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + noise
```

```python
# Initialize Parameters
W = np.random.randn(1, 1)
b = np.zeros((1, 1))
alpha = 0.01  # Learning rate
epochs = 1000  # Number of iterations
m = len(X)  # Number of samples
```

```python
a=X[0]
c= W[0]
d=b
e=a*c+d

y_pred = X.dot(W) + b
f=y_pred[0]
e
if e==f:
    print("Equal")

error = y_pred - y
error.max()

```

```python
# Implement Gradient Descent
for epoch in range(epochs):
    y_pred = X.dot(W) + b  # Predictions
    error = y_pred - y  # Compute error
    
    # Compute gradients
    dW = (1/m) * X.T.dot(error)
    db = (1/m) * np.sum(error)
    
    # Update weights
    W -= alpha * dW
    b -= alpha * db
```

```python
# Plot Results
plt.scatter(X, y, color='orange', label='Training Data')
plt.plot(X, X.dot(W) + b, color='green', label='Linear Regrassion')
plt.xlabel("Features")
plt.ylabel("Labels")
plt.legend()
plt.show()
```

```python
# Evaluate Model using Mean Squared Error
mse = np.mean((X.dot(W) + b - y) ** 2)
print("Mean Squared Error:", mse)
```
